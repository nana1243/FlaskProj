```python
from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from bs4 import BeautifulSoup


import numpy as np
import pandas as pd
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")


# webdriver 설정(Chrome, Firefox 등) - 일반 모드
browser = webdriver.Chrome('webdriver/chromedriver.exe')
```

```python
# 크롬 브라우저 내부 대기
browser.implicitly_wait(5)


# 페이지 이동
browser.get('https://www.ministop.co.kr/MiniStopHomePage/page/event/plus1.do')

```

```python
# 더보기 클릭후, 필요한 부분을 추출하기 위해 pro_list 생성

n=0
for i in range(0,6):
    WebDriverWait(browser, 2).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.inner.wrap.service1 > div.event_plus_list > div > a.pr_more'))).click()
    time.sleep(5)
    n+=1


soup = BeautifulSoup(browser.page_source, "html.parser")
pro_list = soup.select('div.inner.wrap.service1 > div.event_plus_list > ul>li ')
#for v in pro_list:
    #print(v.select('a>p')[0].text.strip())

```

```python
pro_list
```

```python
prod_list=[]
price_list=[]

for v in pro_list:
    prodname=v.select('a>img')[0]["alt"]
    prod_list.append(prodname)
    price=v.select('p>strong')[0].text.strip()
    price_list.append(price)
    
    
```

```python
prod_list
```

```python
price_list
```

```python
df_ministop_oneplusone=pd.DataFrame({"품명":prod_list, "가격":price_list, "행사명":"1+1"})
```
