```python
import numpy as np
import pandas as pd

import os
import urllib.parse as rep
import urllib.request as req
from fake_useragent import UserAgent

import time

# 이미지 다운
# import urllib.request as req

# 이미지 바이트 처리
from io import BytesIO

# 액셀 처리 임포트
import xlsxwriter

# bs4 임포트
from bs4 import BeautifulSoup

# selenium 임포트
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")

browser = webdriver.Chrome('../webdriver/chromedriver.exe')
```

```python
# 브라우저 내부 대기
browser.implicitly_wait(3)

browser.get('http://gs25.gsretail.com/gscvs/ko/products/event-goods')
```

```python
# 1+1 button XPATH
# //*[@id="ONE_TO_ONE"]
WebDriverWait(browser, 2).until(EC.presence_of_element_located((By.XPATH, '//*[@id="GIFT"]'))).click()
```

```python
prod_list=[]
price_list=[]
dum_list=[]
make_list = []
make_list2 = []
# 현재 페이지
cur_page_num = 1

# 크롤링 페이지 수
target_crawl_num =4

while cur_page_num <= target_crawl_num:
    soup = BeautifulSoup(browser.page_source, "html.parser")
    pro_list = soup.select('div.cnt > div.cnt_section.mt50 > div > div > div:nth-child(5) > ul>li')


    # 페이지 번호 출력
    print('****** Current Page : {}'.format(cur_page_num), ' ******')
   
    # 필요 정보 추출(페이지 이동 크롤링)
    for v in pro_list:
        prod_name=v.select('div>p.tit')[0].text
        print(prod_name)
        prod_list.append(prod_name)
        price=v.select('div>p.price>span')[0].text
        print(price)
        price_list.append(price)
        print(v.select('div.dum_txt>p.name')[0].text)
        dum=v.select('div.dum_txt>p.name')[0].text
        dum_list.append(dum)
        
        if len(v.select('div > p.img > img')) > 0:
            
            try:
                prod_img = v.select('div > p.img > img')[0]['src']
                make_list.append(prod_img)
            except Exception as e:
                make_list.append('http://www.kobm.net/images/storage/upload/PRODUCT/PD00000708/20150306174507.PD00000708.01096859.jpg')
                print("Download failed.")
                
        else:
            make_list.append('http://www.kobm.net/images/storage/upload/PRODUCT/PD00000708/20150306174507.PD00000708.01096859.jpg')
        
        
        if len(v.select('div > div.dum_box > div.dum_prd > p.img > img')) > 0:
            try:
                prod_img2 = v.select('div > div.dum_box > div.dum_prd > p.img > img')[0]['src']
                make_list2.append(prod_img2)
            except Exception as e:
                make_list.append('http://www.kobm.net/images/storage/upload/PRODUCT/PD00000708/20150306174507.PD00000708.01096859.jpg')
                print("Download failed.")
        else:
            make_list.append('http://www.kobm.net/images/storage/upload/PRODUCT/PD00000708/20150306174507.PD00000708.01096859.jpg')        

    cur_page_num += 1
    if cur_page_num > target_crawl_num:
        print('Crawling Succeed.')
        break
        time.sleep(1)
    else:
        WebDriverWait(browser, 2).until(EC.presence_of_element_located((By.CSS_SELECTOR, ' div.cnt > div.cnt_section.mt50 > div > div > div:nth-child(7) > div > a.next '))).click()
        time.sleep(2)
```
