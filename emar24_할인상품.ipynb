```python
# replace
prod_list=[]
price_list=[]
# 현재 페이지
cur_page_num = 1

# 크롤링 페이지 수
target_crawl_num =13

while cur_page_num <= target_crawl_num:
    soup = BeautifulSoup(browser.page_source, "html.parser")
    pro_list = soup.select('#regForm > div.section > div.eventProduct > div.tabContArea > ul>li')


    # 페이지 번호 출력
    print('****** Current Page : {}'.format(cur_page_num), ' ******')
   
    # 필요 정보 추출(페이지 이동 크롤링)
    for v in pro_list:
        prod_name=v.select('div > p.productDiv')[0].text.strip()
        print(prod_name)
        prod_list.append(prod_name)
        price=v.select('div > p.price')[0].text.strip()
        print(price)
        price_list.append(price)
        
         

    cur_page_num += 1
    if cur_page_num > target_crawl_num:
        print('Crawling Succeed.')
        break
    else:
        WebDriverWait(browser, 2).until(EC.presence_of_element_located((By.CSS_SELECTOR, 'div.section > div.eventProduct > div.paging > a.next.bgNone > img'))).click()
        time.sleep(2)
```

```python
for i in range(len(price_list)):
    price_list[i]=price_list[i].replace(",","")
    price_list[i]=price_list[i].replace("\xa0원","")
    price_list[i]=price_list[i].split(" →",2)
```

```python
price_list_new=[]
for i in range(len(price_list)):
    price_list_new.append(price_list[i][1])
    
    
```

```python
df_emart24_discount = pd.DataFrame({'품명':prod_list, '가격':price_list_new})
df_emart24_discount['행사명'] = '가격할인'
df_emart24_discount
```
