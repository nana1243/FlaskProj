```python
# 세븐일레븐 1+1 크롤링

from selenium import webdriver
import time
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup

import urllib.request as req
# 이미지 바이트 처리
from io import BytesIO
import os
import urllib.parse as rep
import urllib.request as req
from fake_useragent import UserAgent
import numpy as np
import pandas as pd



from urllib.error import URLError, HTTPError

```

```python
chrome_options = Options()
chrome_options.add_argument("--headless")

# webdriver 설정(Chrome, Firefox 등) - Headless 모드
# browser = webdriver.Chrome(
#             '../webdriver/chromedriver.exe', 
#             options=chrome_options)

# webdriver 설정(Chrome, Firefox 등) - 일반 모드
browser = webdriver.Chrome('webdriver/chromedriver.exe')

# 크롬 브라우저 내부 대기
browser.implicitly_wait(5)

# 브라우저 사이즈
browser.set_window_size(1920, 1280)  # maximize_window(), minimize_window()

# 페이지 이동
browser.get('https://www.7-eleven.co.kr/product/presentList.asp')
time.sleep(1)
# 1차 페이지 내용
# print('Before Page Contents : {}'.format(browser.page_source))
```

```python
time.sleep(1)
body = browser.find_element_by_tag_name("body")
num_of_pagedowns = 20
browser.find_element_by_xpath('//*[@id="listUl"]/li[15]/a').click()
while num_of_pagedowns :
    body.send_keys(Keys.PAGE_DOWN)
    time.sleep(1)
    num_of_pagedowns -=1
    try :
        browser.find_element_by_xpath('//*[@id="moreImg"]/a').click()
    except:
        None
        
        
```

```python
# bs4 초기화
soup = BeautifulSoup(browser.page_source, "html.parser")

# 소스코드 정리
# print(soup.prettify())

# 메인 상품 리스트 선택
# pro_list = soup.select('div.main_prodlist.main_prodlist_list > ul > li')
pro_list = soup.select('div.img_list.img_list_01 > ul > li > div')
#productListArea > div.main_prodlist.main_prodlist_list
#productListArea > div.main_prodlist.main_prodlist_list
# 상품 리스트 확인
print(pro_list)
```

```python
# Hearder 정보 초기화
opener = req.build_opener()
# User-Agent 정보
opener.addheaders = [('User-agent', UserAgent().ie)]
# Header 정보 삽입
req.install_opener(opener)
```

```python
savePath = "C:/Users/전산실/conv_se_1+1/"
try :
   # 기존 폴더가 잇는지 체크
   if not (os.path.isdir(savePath)):
        # 없으면 폴더 생성
        os.makedirs(os.path.join(savePath))
except OSError as e:
    # 에러 내용
    print("folder creation failed!")
    print("folder name : {}".format(e.filename))
    
    # 런타임 에러 raise
    raise RuntimeError('System Exit!')

else:
    # 폴더 생성 정상일 경우 실행
    print('folder is created!')
```

```python
# 이미지 다운로드
make_list = []
for v in pro_list:
    make_list.append(v.select('div.pic_product > img')[0]['src'])

for i in range(0,len(make_list)) :

    fullFileName = os.path.join(savePath, savePath + str(i) + '.png')
    req.urlretrieve('http://www.7-eleven.co.kr'+ make_list[i], fullFileName)


# # 이미지 번호를 붙여주면서 다운로드
# for i, pro_list in enumerate(pro_list, 1) : 
#     fullFileName = os.path.join(savePath, savePath + str(i) + '.png')
#     madeUrl = 'http://www.7-eleven.co.kr'+pro('div.pic_product > img')[0]['src']
#     req.urlretrieve(+pro_list['src'], fullFileName)
```

```python
name = []
price = []


for v in pro_list:
    prod_name = v.select('div.infowrap > div')[0].text.strip()
    name.append(prod_name)
    prod_price = v.select('span')[0].text.strip()
    price.append(prod_price)

print(name)
print(price)
```

```python
df_seven11_oneplusone= pd.DataFrame({'품명':name, '가격':price, '행사명':'1+1'})
```
